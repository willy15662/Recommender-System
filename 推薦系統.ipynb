```python
# 使用 read_parquet 加载parquet文件
import pandas as pd
import numpy as np
from pandas import read_parquet
data = read_parquet("/content/drive/MyDrive/Colab Notebooks/神經網路/dataset/label_test_source.parquet")
# data1 = read_parquet("/content/drive/MyDrive/Colab Notebooks/神經網路/dataset/label_train_source.parquet")
# data2 = read_parquet("/content/drive/MyDrive/Colab Notebooks/神經網路/dataset/label_train_target.parquet")
# data3 = read_parquet("/content/drive/MyDrive/Colab Notebooks/神經網路/dataset/meta_song.parquet")
# data4 = read_parquet("/content/drive/MyDrive/Colab Notebooks/神經網路/dataset/meta_song_composer.parquet")
# data5 = read_parquet("/content/drive/MyDrive/Colab Notebooks/神經網路/dataset/meta_song_genre.parquet")
# data6 = read_parquet("/content/drive/MyDrive/Colab Notebooks/神經網路/dataset/meta_song_lyricist.parquet")
# data7 = read_parquet("/content/drive/MyDrive/Colab Notebooks/神經網路/dataset/meta_song_producer.parquet")
# data8 = read_parquet("/content/drive/MyDrive/Colab Notebooks/神經網路/dataset/meta_song_titletext.parquet")
```

```python
df = data
print(df.count())
df.head(30)
```

```python
# Group by 'session_id' and 'song_id' and count occurrences
grouped_df = df.groupby(['session_id', 'song_id']).size().reset_index(name='count')

# Sort within each group by count in descending order
sorted_df = grouped_df.sort_values(['session_id', 'count'], ascending=[True, False])

# Extract the top five songs for each session
top_five_songs = sorted_df.groupby('session_id').head(5)

# Create a new column 'rank' indicating the rank of each song within each session
top_five_songs['rank'] = top_five_songs.groupby('session_id').cumcount() + 1

# Pivot the data to create a new DataFrame with 'session_id' and top five songs
df_pivoted = top_five_songs.pivot(index='session_id', columns='rank', values='song_id')

# Rename the columns to 'top1', 'top2', 'top3', 'top4', 'top5'
df_pivoted.columns = [f'top{i}' for i in range(1, 6)]

# Reset the index to include 'session_id' as a regular column
df_pivoted.reset_index(inplace=True)
```

```python
df_pivoted.head(10)
```

```python
# 假設 df_pivoted 是你的資料
# 使用 bfill（向後填補）來填補缺失值
df_pivoted[['top2', 'top3', 'top4', 'top5']] = df_pivoted[['top2', 'top3', 'top4', 'top5']].fillna(method='bfill', axis=1)

# 建立一個 fill_value 欄位，根據不同情況選擇 'top1'、'top2' 等欄位的值
fill_value = df_pivoted.apply(lambda row: row[['top1', 'top2', 'top3', 'top4', 'top5']].dropna().tolist() if any(row[['top1', 'top2', 'top3', 'top4', 'top5']].notna()) else [], axis=1)

# 使用 fill_value 來填補缺失值
df_pivoted[['top2', 'top3', 'top4', 'top5']] = df_pivoted[['top2', 'top3', 'top4', 'top5']].apply(lambda col: col.fillna(fill_value), axis=0)
```

```python
# replacing null values with the value from 'top1' column:
df_pivoted.fillna({'top2': df_pivoted['top1'], 'top3': df_pivoted['top1'], 'top4': df_pivoted['top1'], 'top5': df_pivoted['top1']}, inplace=True)

# Save the resulting DataFrame to a new CSV file
df_pivoted.to_csv('output.csv', index=False)
#print(df_pivoted)
```
